```python
kernel=['linear','rbf','sigmoid','poly']
def SVM_function(kernal,gamm,C,samples,j):
    SVM=svm.SVC(kernel=kernal,gamma=gamm,C=C,max_iter=1000)
    X_train,X_test,y_train,y_test=samples[j]
    SVM.fit(X_train,y_train)
    y_pred=SVM.predict(X_test)
    return accuracy_score(y_pred,y_test)
```
